## SAS NOTLARI

### Kütüphane Oluşturma
libname DENEME 'kurmak istediğiniz adres' ;

Örnek:
    libname DENEME '/home/u59204182/sasuser.v94' ; komutuyla kütüphanemizi kurabiliriz.
    
Kütüphanenin kurulup kurulmadığını Libraries sekmesinde MyLibraries sekmesine tıklayıp oluşup oluşmadığını görebilirsiniz.


Görüldüğü Üzere Kendi Kütüphanemizi DENEME adında oluşturmuş olduk.

SAS'ta iki önemli statament vardır bunlar **data**; tabloyu oluşturur ve **run**; kodu çalıştırır, bunların arasına yazdığımız kod parçaçıkları çalışır.

### Tablo Oluşturma


İlk tablomuzu oluşturalım notlarda data tablo oluşturur demiştik. Data'nın altında tablonun nereye yapılacağı ve elemanlarını yazıyoruz.



Tablomuz oluşmuş oldu. Tablomuzda a ve a'nın değeri yazmaktadır görüldüğü üzere



WORK kütüphanesinin altına kaydedilen tüm bilgiler programı kapadığımız zaman silinir. Eğer tabloyu kalıcı olarak eklemek istiyorsa kendi DENEME kütüphanemize kaydetmemiz gerekir.



Kalıcı Tablomuzu oluşturduk.

**set**: bir tabloyu başka bir tablonun üzerine yazmak istersek, tabloyu olduğu gibi yazmak istersek
yani herhangi bir şekilde tablo kullanmak istersek set komutunu kullanırız.
